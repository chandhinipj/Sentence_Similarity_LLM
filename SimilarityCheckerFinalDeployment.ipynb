```python
%%writefile app.py
import streamlit as st
import io
import dill

filePath = open('/content/stsb-distilroberta-base-v2_PredictFile', 'rb')
predict = dill.loads(filePath.read())

def main():
    # Set Streamlit page title and config
    st.set_page_config(
        page_title='Text Similarity Checker',
        page_icon=':pencil:',  # Add your desired page icon
        layout='centered'  # Change the layout if desired
    )
    st.set_option('deprecation.showfileUploaderEncoding', False)
    # Set page background image using CSS
    st.markdown(
        """
        <style>
        body {
            background-color: #f4f4f4;
            background-image: url('https://miro.medium.com/v2/resize:fit:1400/1*NTPw3UFr_rtUSz8gfjaQQg.png');
            background-size: cover;
            background-repeat: no-repeat;
            background-attachment: fixed;
        }
        </style>
        """,
        unsafe_allow_html=True
    )

    # Streamlit components
    st.title('Text Similarity Checker')

    sentence1 = st.text_input('Enter First Sentence', value='')
    sentence2 = st.text_input('Enter Second Sentence', value='')

    if st.button('Check Similarity', key='process_button'):
        if sentence1 and sentence2:
            sentence_scoring = predict(sentence1, sentence2)
            st.markdown(f'<p style="text-align: center; font-size: 24px;">Similarity Score: {sentence_scoring}</p>', unsafe_allow_html=True)
        else:
            st.warning('Please provide both sentences for comparison.')

if __name__ == '__main__':
    main()
```
